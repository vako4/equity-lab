```python
import pandas as pd
import sqlite3
from analytics import summary_stats

# Load prices from your SQLite DB
conn = sqlite3.connect("equity_lab.db")
prices_long = pd.read_sql(
    "SELECT date, ticker, close FROM prices ORDER BY date",
    conn,
    parse_dates=["Date"]
)
conn.close()

# Pivot from long format (one row per date+ticker) to wide format (one column per ticker)
prices_wide = prices_long.pivot(index="Date", columns="Ticker", values="Close")

prices_wide.head()
```

```python
summary_stats(prices_wide).sort_values("Sharpe", ascending=False)
```
